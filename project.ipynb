Import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Read the dataset

```python
df = pd.read_csv("./Data/2008.csv", encoding='ISO-8859-1', low_memory=False)
df.index.name="Id"
df
```

Get hours from the data and calculate mean for each.

```python
df['Hour'] = np.floor(df.CRSDepTime/100)
df = df.groupby(['Hour'])['DepDelay'].mean().reset_index()
df
```

Display graph 

```python
ax = df.plot.bar(x='Hour', y='DepDelay', rot=0, figsize=[14.70, 8.27])
plt.title("Distribution of delay over the hours day", fontsize = 14, weight = "bold")
plt.ylabel('Delay', fontsize = 10, weight = "bold")
plt.show()
```

Interpretation:<br>
It can be seen that night and morning flights tend to be the most on-time.<br>
Flights scheduled between 12:00 am. and 10 am. generally have delays under 10 minutes.<br>
The best times for departure are 5:00 am and 6:00 am, then the delay is almost imperceptible.<br>
After that almost uninterrupted (at 8 pm the delay slightly decreases) upward trend can be observed with a peak at 11 am.
