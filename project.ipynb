Import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Read the dataset

```python
df = pd.read_csv("./Data/2008.csv", encoding='ISO-8859-1', low_memory=False)
df.index.name="Id"
df
```

Get hours from the data

```python
df['Hour'] = np.floor(df.CRSDepTime/100)
df = df.groupby(['Hour'])['DepDelay'].mean().reset_index()
df
```

Display graph 

```python
ax = df.plot.bar(x='Hour', y='DepDelay', rot=0, figsize=[14.70, 8.27])
plt.title("Distribution of delay over the hours day", fontsize = 14, weight = "bold")
plt.ylabel('Delay', fontsize = 10, weight = "bold")
plt.show()
```

Interpretation:<br>
It can be seen that the best time for departure are 5am and 6am.<br>
After that almost uninterrupted (at 8p.m. the delay slightly decreases) upward trend can be observed with a peak at 11pm.
