Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Read the dataset

```python
df = pd.read_csv("./Data/2008.csv", encoding='ISO-8859-1', low_memory=False)
df.index.name="Id"
df
```

Group by day of week and mean delay

```python
df = df.groupby('DayOfWeek')['DepDelay'].mean().reset_index()
df
```

Map 

```python
days = {1:'Monday', 2:'Tuesday', 3:'Wednesday', 4:'Thursday', 5:'Friday', 6:'Saturday', 7:'Sunday'}
df['DayOfWeek'] = df['DayOfWeek'].map(days)
```

Display graph 

```python
ax = df.plot.bar(x='DayOfWeek', y='DepDelay', rot=0, figsize=[14.70, 8.27])
plt.title("Distribution of delay over the week days", fontsize = 14, weight = "bold")
plt.ylabel('Delay', fontsize = 10, weight = "bold")
plt.xlabel('Day of Week', fontsize = 10, weight = "bold")
plt.show()
```

Interpretation:<br>
Looking at the resulting graph, it can be said that Friday is the worst day to have a flight.<br>
On the other hand Wednesday is the best option if we want to avoid a delay.
