```python
df = pd.read_csv("./DataExpo2009/2001.csv", encoding='ISO-8859-1', low_memory=False)
```

```python
# Give a unique row number as "Id".
df.index.name="Id"
df.reset_index()
func = dict(Cancelled = 'sum')
df = df.groupby(['Month']).agg(func).reset_index()
df
```

```python
#sns.set_theme(style="whitegrid")
sns.barplot(data = df, x = 'Month', y = 'Cancelled', palette = sns.color_palette('hls', 1))
```
